```python
import numpy as np

def to_sequences(seq_size,obs):
    x=[]
    y=[]
    for i in range(len(obs)-seq_size-1):
        window = obs[i:(i+seq_size)]
        after_window = obs[i+seq_size]
        window =[[x] for x in window]
        x.append(window)
        y.append(after_window)
    return np.array(x), np.array(y)

SEQ_SIZE = 10

x_train1, y_train1 = to_sequences(SEQ_SIZE, df_data_train1)
x_test1, y_test1 = to_sequences(SEQ_SIZE, df_data_test1)

x_train2, y_train2 = to_sequences(SEQ_SIZE, df_data_train2)
x_test2, y_test2 = to_sequences(SEQ_SIZE, df_data_test2)

x_train3, y_train3 = to_sequences(SEQ_SIZE, df_data_train3)
x_test3, y_test3 = to_sequences(SEQ_SIZE, df_data_test3)

x_train4, y_train4 = to_sequences(SEQ_SIZE, df_data_train4)
x_test4, y_test4 = to_sequences(SEQ_SIZE, df_data_test4)

x_train5, y_train5 = to_sequences(SEQ_SIZE, df_data_train5)
x_test5, y_test5 = to_sequences(SEQ_SIZE, df_data_test5)

x_train6, y_train6 = to_sequences(SEQ_SIZE, df_data_train6)
x_test6, y_test6 = to_sequences(SEQ_SIZE, df_data_test6)

x_train7, y_train7 = to_sequences(SEQ_SIZE, df_data_train7)
x_test7, y_test7 = to_sequences(SEQ_SIZE, df_data_test7)

x_train8, y_train8 = to_sequences(SEQ_SIZE, df_data_train8)
x_test8, y_test8 = to_sequences(SEQ_SIZE, df_data_test8)

x_train9, y_train9 = to_sequences(SEQ_SIZE, df_data_train9)
x_test9, y_test9 = to_sequences(SEQ_SIZE, df_data_test9)

x_train10, y_train10 = to_sequences(SEQ_SIZE, df_data_train10)
x_test10, y_test10 = to_sequences(SEQ_SIZE, df_data_test10)

x_train11, y_train11 = to_sequences(SEQ_SIZE, df_data_train11)
x_test11, y_test11 = to_sequences(SEQ_SIZE, df_data_test11)
```

```python
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense, Embedding
from keras.layers import LSTM
from keras.datasets import imdb
import numpy as np
from sklearn import metrics
from keras.callbacks import History 
history = History()

def create_model():
    model = Sequential()
    model.add(LSTM(50, return_sequences = True, input_dim = 1))
    model.add(LSTM(50, return_sequences = False))
    model.add(Dense(1))
    model.compile(loss = 'mae', optimizer = 'adam')
    return model

m = create_model()

m.fit(x_train1, y_train1, validation_data=[x_test1, y_test1], verbose = 2, epochs = 300, callbacks=[history])
pred1 = m.predict(x_test1)
score1 = np.sqrt(metrics.mean_squared_error(pred1, y_test1))
hist1 = history.history

m.fit(x_train2, y_train2, validation_data=[x_test2, y_test2], verbose = 2, epochs = 300, callbacks=[history])
pred2 = m.predict(x_test2)
score2 = np.sqrt(metrics.mean_squared_error(pred2, y_test2))
hist2 = history.history

m.fit(x_train3, y_train3, validation_data=[x_test3, y_test3], verbose = 2, epochs = 300, callbacks=[history])
pred3 = m.predict(x_test3)
score3 = np.sqrt(metrics.mean_squared_error(pred3, y_test3))
hist3 = history.history

m.fit(x_train4, y_train4, validation_data=[x_test4, y_test4], verbose = 2, epochs = 300, callbacks=[history])
pred4 = m.predict(x_test4)
score4 = np.sqrt(metrics.mean_squared_error(pred4, y_test4))
hist4 = history.history

m.fit(x_train5, y_train5, validation_data=[x_test5, y_test5], verbose = 2, epochs = 300, callbacks=[history])
pred5 = m.predict(x_test5)
score5 = np.sqrt(metrics.mean_squared_error(pred5, y_test5))
hist5 = history.history

m.fit(x_train6, y_train6, validation_data=[x_test6, y_test6], verbose = 2, epochs = 300, callbacks=[history])
pred6 = m.predict(x_test6)
score6 = np.sqrt(metrics.mean_squared_error(pred6, y_test6))
hist6 = history.history

m.fit(x_train7, y_train7, validation_data=[x_test7, y_test7], verbose = 2, epochs = 300, callbacks=[history])
pred7 = m.predict(x_test7)
score7 = np.sqrt(metrics.mean_squared_error(pred7, y_test7))
hist7 = history.history

m.fit(x_train8, y_train8, validation_data=[x_test8, y_test8], verbose = 2, epochs = 300, callbacks=[history])
pred8 = m.predict(x_test8)
score8 = np.sqrt(metrics.mean_squared_error(pred8, y_test8))
hist8 = history.history

m.fit(x_train9, y_train9, validation_data=[x_test9, y_test9], verbose = 2, epochs = 300, callbacks=[history])
pred9 = m.predict(x_test9)
score9 = np.sqrt(metrics.mean_squared_error(pred9, y_test9))
hist9 = history.history

m.fit(x_train10, y_train10, validation_data=[x_test10, y_test10], verbose = 2, epochs = 300, callbacks=[history])
pred10 = m.predict(x_test10)
score10 = np.sqrt(metrics.mean_squared_error(pred10, y_test10))
hist10 = history.history

m.fit(x_train11, y_train11, validation_data=[x_test11, y_test11], verbose = 2, epochs = 300, callbacks=[history])
pred11 = m.predict(x_test11)
score11 = np.sqrt(metrics.mean_squared_error(pred11, y_test11))
hist11 = history.history
```
